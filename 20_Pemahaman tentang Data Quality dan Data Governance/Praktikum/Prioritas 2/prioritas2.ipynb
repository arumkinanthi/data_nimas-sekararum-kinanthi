#### Import Library

```python
import pandas as pd
```

#### Membaca file .csv

```python
df = pd.read_csv("tx_data.csv")
df.head(10)
```

#### Mengubah format nomor telepon dengan awalan +62 di depan

```python
def parsedata(x):
   x = str(x).replace('-', '')
   x = '+62' + x
   return x
```

```python
df["phone_number"] = df["phone_number"].apply(lambda x : parsedata(x))
df.head(10)
```

#### Mengubah nilai transaksi (transaction_amount) dengan format seperti berikut: Rp 12.000

```python
def format_transaksi(amount):
   amount_str = str(amount)
   if len(amount_str) <= 3:
      return "Rp " + amount_str  #jika nilai kurang dari atau sama dengan 3 digit, langsung menambahkan "Rp"
   else:
      formatted = "Rp " + amount_str[:-3] + "." + amount_str[-3:]  #untuk menambahkan titik di tengah nilai
      return formatted
```

```python
df["transaction_amount"] = df["transaction_amount"].apply(lambda x : format_transaksi(x))
df.head(10)
```

#### Mengambil data transaksi dengan status sama dengan success

```python
def transform(df, columnname, condition):
   filtered_df = df[(df[columnname] == condition)]
   return filtered_df
```

```python
transformed_df = transform(df, "transaction_status", "success")
transformed_df.head(10)
```

#### Menyimpan hasil akhir data cleaning di dalam DataFrame

```python
transformed_df = transformed_df.reset_index(drop=True)
data_frame = transformed_df
data_frame
```
