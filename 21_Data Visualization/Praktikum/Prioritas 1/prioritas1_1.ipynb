#### Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Membaca file .csv

```python
df = pd.read_csv("electronic_shop.csv")
df.head(10)
```

#### Line plot jumlah nilai dari transaksi per hari

```python
df.set_index("transaction_id")
sum_transaction_date = df.groupby(["transaction_date"]).sum(numeric_only=True)
transaction_dates = sum_transaction_date.index.get_level_values(0).tolist()
amount_sum = sum_transaction_date["transaction_amount"]

plt.figure(figsize=(15, 7))
plt.title("Transaction Amount Sum per Day")
plt.xlabel("Transaction Dates")
plt.ylabel("Total")
plt.plot(transaction_dates, amount_sum)
plt.show()
```

#### Bar plot jumlah customer per hari

```python
customer_counts_per_days = df.groupby(["transaction_date"]).count()
transaction_dates = customer_counts_per_days.index.get_level_values(0).tolist()
customers = customer_counts_per_days["customer_name"]

plt.title("Total Customers per Day")
plt.xlabel("Transaction Dates")
plt.ylabel("Customers")
plt.bar(transaction_dates, customers)
plt.show()
```

#### Scatter plot jumlah nilai dari transaksi per hari

```python
plt.title("Transaction Amount Sum per Day")
plt.xlabel("Transaction Date")
plt.ylabel("Transaction Amount")
sns.scatterplot(x="transaction_date", y="transaction_amount", data=df, hue="payment_method")
```
