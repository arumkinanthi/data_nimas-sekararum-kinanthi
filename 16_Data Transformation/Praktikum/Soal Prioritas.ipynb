## Import Library

```python
import pandas as pd
import requests
import validators
```

## Ekstrak data ke dalam dataframe

```python
df = pd.read_csv("house_listings.csv")
df.head()
```

#### Mengubah dan merapikan data harga dan harga per meter persegi

```python
df['price'] = df['price'].astype('string')
df['price'] = df['price'].str.replace(' ', '')
df['price'].fillna('0', inplace=True)
df['price'] = df['price'].astype(int)

df['price_1m2'] = df['price_1m2'].astype('string')
df['price_1m2'] = df['price_1m2'].str.replace('AZN/m²', '')
df['price_1m2'] = df['price_1m2'].str.replace(' ', '')
df['price_1m2'].fillna('0', inplace=True)
df['price_1m2'] = df['price_1m2'].astype(int)

df.head()
```

## Normalisasi harga rumah dan harga per meter persegi menggunakan teknik Min-Max
#### 1. Normalisasi Harga Rumah

```python
max_price = df["price"].max()
min_price = df["price"].min()

for x in range(len(df['price'])):
   xvalue = df['price'][x]
   df['price'][x] = (xvalue - min_price)/(max_price - min_price)
print(df['price'])
```

#### 2. Harga per meter persegi

```python
max_price_1m2 = df["price_1m2"].max()
min_price_1m2 = df["price_1m2"].min()


for x in range(len(df['price_1m2'])):
   xvalue = df['price_1m2'][x]
   df['price_1m2'][x] = (xvalue - min_price_1m2)/(max_price_1m2 - min_price_1m2)
df
```

## One-hot Encoding pada Kolom 'category', 'title_deed', 'repair', dan 'mortgage'

```python
df['category'] = df['category'].astype('string')
df['category'] = df['category'].str.replace('Köhnə tikili', 'Köhnə_tikili')
df['category'] = df['category'].str.replace('Yeni tikili', 'Yeni_tikili')
df['category'].fillna('NULL_tikili', inplace=True)

df['title_deed'] = df['title_deed'].astype('string')
df['title_deed'].fillna('NULL', inplace=True)

df['repair'] = df['repair'].astype('string')
df['repair'].fillna('NULL', inplace=True)

df['mortgage'] = df['mortgage'].astype('string')
df['mortgage'].fillna('NULL', inplace=True)

encoded_df = pd.get_dummies(df, columns=['category', 'title_deed', 'repair', 'mortgage'], dtype=int)
encoded_df
```

## Aggregasi: Menghitung rata-rata, median, dan modus dari harga rumah berdasarkan jumlah kamar dan tipe bangunan

```python
grouped_df = df.groupby(['floor', 'category'])

mean_price = grouped_df['price'].mean()
median_price = grouped_df['price'].median()
mode_price = grouped_df['price'].apply(lambda x: x.mode().iloc[0])

print("Rata-rata Harga Rumah:")
print(mean_price)
```

```python
print("Median Harga Rumah:")
print(median_price)
```

```python
print("Modus Harga Rumah:")
print(mode_price)
```

## Mengidentifikasi dan menginput nilai yang hilang pada kolom 'area' dan 'floor'

```python
df['area'] = df['area'].astype('string')
df['area'].fillna('0 m²', inplace=True)

df['floor'] = df['floor'].astype('string')
df['floor'].fillna('0 / 0', inplace=True)
df
```

## Menggunakan metode IQR untuk mengidentifikasi dan menangani outlier pada harga rumah

```python
quartile1 = df['price'].quantile(0.25)
quartile3 = df['price'].quantile(0.75)
iqr = quartile3 - quartile1

lower_bound = quartile1 - 1.5 * iqr
upper_bound = quartile3 + 1.5 * iqr

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]

# mengganti nilai outlier dengan nilai median
median = df['price'].median()
df.loc[(df['price'] < lower_bound) | (df['price'] > upper_bound), 'price'] = median

print("Outlier:")
print(outliers)
print("\nDataFrame setelah penanganan outlier:")
df
```

## Memverifikasi dan memvalidasi keakuratan data pada kolom 'address' dan 'url'

```python
df['address'] = df['address'].astype('string')
df['address'].fillna('NULL', inplace=True)

df['url'] = df['url'].astype('string')
df['url'].fillna('NULL', inplace=True)
```

```python
def validate_address(address):
    # panjang alamat harus lebih dari 5 karakter
    if len(address) > 5:
        return True
    else:
        return False

df['address_is_valid'] = df['address'].apply(validate_address)

def validate_url(url):
    if pd.notnull(url) and validators.url(url):
        try:
            # mengecek ketersediaan URL dengan mengirim request
            response = requests.head(url)
            if response.status_code == 200:
                return True
            else:
                return False
        except requests.ConnectionError:
            return False
    else:
        return False

df['url_is_valid'] = df['url'].apply(validate_url)

print("DataFrame setelah validasi:")
print(df)
```
