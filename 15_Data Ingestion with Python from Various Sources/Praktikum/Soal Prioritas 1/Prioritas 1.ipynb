### Library Import

```python
import pandas as pd
from IPython.display import display
from pymysql import connect
import requests
```

## Ingest Data from csv

```python
df_csv = pd.read_csv("transactions.csv")
display(df_csv.head())
display(df_csv.tail())
df_csv.to_parquet("transactions_from_csv.parquet")
```

## Ingest Data from Excel

```python
df_excel = pd.read_excel("products.xlsx")
display(df_excel.head())
display(df_excel.tail())
df_excel.to_parquet("products_from_xlsx.parquet")
```

### Ingest Data from Database (MySQL)

```python
connection = connect(
   host="127.0.0.1",
   db="customer",
   user="root",
   password="@Kinan673",
)

cursor = connection.cursor()
cursor.execute("SELECT * FROM customers")
results = cursor.fetchall()

df_mysql = pd.DataFrame(results, columns = ["id", "name", "email", "address"])
display(df_mysql.head())
display(df_mysql.tail())
df_mysql.to_parquet("customers_from_mysql.parquet")
```

## Ingest Data from APIs

```python
url = 'https://fakestoreapi.com/products'

df_api = pd.DataFrame(requests.get(url).json())
display(df_api.head())
display(df_api.tail())
df_api.to_parquet("products_from_api.parquet")
```
